# Count of medical Condition

```python
medical_condition_counts = df['Medical Condition'].value_counts()
print(medical_condition_counts)
```

```python
col= ['Medical Condition','Gender']
grouped = df.groupby(col)['Medical Condition'].count().reset_index(name='count')
print(grouped)



```

#Count of test results

```python
test_results = df['Test Results'].value_counts()
print(test_results)
```

#pie chart for test results


```python
plt.title("Test Results in a bar graph")
plt.pie(test_results,labels= test_results.index,autopct = '%1.2f%%',)
plt.show()
```

#bar graph for medical condtion

```python
plt.title("medical condition in a bar graph")
plt.xlabel("Medical Condition")
plt.ylabel("count")
condition_count = df['Medical Condition'].value_counts()
condition_count.plot(kind = 'bar',color = 'blue')
plt.show()
```

#pie chart for medical condition

```python
plt.title("medical condition in a pie chart")
condition_count = df['Medical Condition'].value_counts()
condition_count.plot(kind = 'pie', autopct = '%1.2f%%')
plt.show()
```

#medical condition &test results with heatmap

```python
crosstab  = pd.crosstab(df['Medical Condition'],df['Test Results'])
plt.title("crosstab of medical condition and test results")
plt.xlabel('test reults')
plt.ylabel('medical condition')
sns.heatmap(crosstab,annot = True)
plt.show()
```

```python
crosstab  = pd.crosstab(df['Medical Condition'],df['Test Results'])

crosstab.plot(kind = 'barh')
```
